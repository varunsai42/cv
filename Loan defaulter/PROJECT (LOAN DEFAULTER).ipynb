```python
a=a.drop(['months_loan_duration','amount'],axis=1)
a
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
for x in a.columns:
    if a[x].dtype=='object':# apply for columns with ctegorical strings
        a[x]=le.fit_transform(a[x])
    
```

```python
a.head()
```

```python
x=a.drop('default',axis=1)
y=a['default']
```

```python
# splitting
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# desion tree
from sklearn import tree
dt=tree.DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
# score 
dt.score(x_train,y_train)
```

```python
dt.score(x_test,y_test)
```

# this is a over fittting model

```python
# reduce over fitting(regularization)
dtr=tree.DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=1)
dtr.fit(x_train,y_train)
```

```python
print(dtr.score(x_train,y_train))
print(dtr.score(x_test,y_test))
```

```python
from sklearn.ensemble import BaggingClassifier
bgcl=BaggingClassifier()
bgcl.fit(x_train,y_train)
```

```python
print(bgcl.score(x_train,y_train))
print(bgcl.score(x_test,y_test))
```
